```python
from postprocessing.libs.spectrum import specanalysis
import matplotlib.pyplot as plt


S = specanalysis()

linewidth = 2.5
plt.rcParams.update({'font.size':'20',
                     'font.family':'serif',
                     'font.weight':'bold',
                     'lines.linewidth':linewidth,
                     'text.usetex':True})
```

# Spectra using binning

```python
# Spectra using binning
# Ti and Tf are the inital and final time for which averaging needs to be done
k, kei_spec, kec_spec = S.ke_spec(Ti = 0, Tf = 0, type='bin')
```

```python
# Plot
plt.loglog(k, kec_spec)
plt.ylim(bottom = 1e-6)
plt.savefig(S.op/"compspec_bin.jpeg", dpi=300, bbox_inches='tight')
```

```python
plt.loglog(k, kei_spec)
plt.ylim(bottom = 1e-6)
plt.savefig(S.op/"incompspec_bin.jpeg", dpi=300, bbox_inches='tight')
```

```python
#Parspec Using Binning
k, parspec = S.par_spec(Ti = 0)
plt.loglog(k, parspec)
plt.ylim(bottom = 1e-6)
plt.savefig(S.op/"parspec_bin.jpeg", dpi=300, bbox_inches='tight')
```

# Spectra using angle averaged Weiner-Khinchin theorem

```python
k, kei_spec, kec_spec = S.ke_spec(Ti = 0, Tf = 0, type='resolved', N = 1000)
```

```python
# Plot
plt.loglog(k, kec_spec)

plt.ylim(bottom = 1e-6)
plt.savefig(S.op/"res_compspec_bin.jpeg", dpi=300, bbox_inches='tight')
```

```python
plt.loglog(k, kei_spec)
plt.ylim(bottom = 1e-6)
plt.savefig(S.op/"res_incompspec_bin.jpeg", dpi=300, bbox_inches='tight')
```

# Fluxes

### (a) Kinetic Energy Fluxes

```python
k, kei_flux, kec_flux = S.ke_flux(Ti = 3, Tf = 3, N = 1000)
```

```python
plt.plot(k, kei_flux)
# plt.ylim(bottom = 1e-6)
plt.xscale('log')
plt.yscale('symlog')
plt.xlim(right = 3e1)
plt.savefig(S.op/"incompflux.jpeg", dpi=300, bbox_inches='tight')
```

```python
plt.plot(k, kec_flux)
plt.xscale('log')
plt.yscale('symlog')
plt.xlim(right = 3e1)
plt.savefig(S.op/"compflux.jpeg", dpi=300, bbox_inches='tight')
```

### (b) Par Flux

```python
k, par_flux = S.par_flux(Ti = 5, Tf = 5)
```

```python
plt.plot(k, par_flux)
plt.xscale('log')
plt.yscale('symlog')
plt.xlim(right = 3e1)
plt.savefig(S.op/"parflux.jpeg", dpi=300, bbox_inches='tight')
```
